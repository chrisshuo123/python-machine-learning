```python
import pandas as pd

train = pd.read_csv(r"dataset/train.csv")
```

Run This Before Dropping

```python
# Define threshold (e.g., 50% missing)
threshold = 0.5
over = train.columns[train.isnull().mean() > threshold]

print("Columns to be deleted:", over.tolist())
print("Number of columns to delete:", len(over))
```

Handle remaining missing values in other columns

```python
# Cek apakah masih ada kolom dengan missing values setelah dihapus
print("Missing values setelah drop columns:")
print(train.isnull().sum().sort_values(ascending=False))
```

Cek Kolom yang akan dihapus

```python
# Cek kolom yang akan dihapus
print("Kolom yang akan dihapus:")
print(over.tolist())
print(f"\nJumlah kolom yang akan dihapus: {len(over)}")
print(f"Persentase missing per kolom:")
for col in over:
    print(f"{col}: {train[col].isnull().mean()*100:.1f}%")
```

Remove the Missing Values

```python
# Menghapus kolom dengan terlalu banyak nilai yang hilang
df = train.drop(columns=over)
```

```python
# Cek hasil penghapusan
print("Shape setelah drop:", df.shape)
print("Kolom yang tersisa:", df.columns.tolist())
```

memverifikasi tidak ada kolom over yang tersisa:

```python
# Cek missing values di dataframe baru
print(df.isnull().sum().sort_values(ascending=False))
```

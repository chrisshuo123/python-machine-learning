### 2. Run Your Code

```python
from scipy import optimize

def f(x):
    return x**2 + 5*x + 6

result = optimize.minimize(f, 0)
print(result)
```

<b>Expected Output:</b><br>
fun: -0.25<br>
 hess_inv: array([[0.5]])<br>
      jac: array([0.])<br>
  message: 'Optimization terminated successfully.'<br>
     nfev: 12<br>
      nit: 4<br>
     njev: 6<br>
   status: 0<br>
  success: True<br>
        x: array([-2.5])

### Understanding the Result
<ul>
    <li><b>x: -2.5</b> → The minimum point (where derivative = 0)</li>
    <li><b>fun: -0.25</b> → Minimum value of f(x) = (-2.5)² + 5(-2.5) + 6 = -0.25</li>
    <li><b>success: True</b> → Optimization worked</li>
</ul>

### To Get Cleaner Output

```python
print(f"Minimum at x = {result.x[0]:.2f}")
print(f"Minimum value = {result.fun:.2f}")
```

#### Visualize the Result

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 1, 100)
y = x**2 + 5*x + 6

plt.plot(x, y, label='f(x) = x² + 5x + 6')
plt.scatter(result.x, result.fun, color='red', s=100, label='Minimum')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

The code finds the minimum of the quadratic function starting from x=0. Works perfectly!
